## Data visualization
The following code is meant to read a .txt file containing the time series outcome of a single-instance run of a simulation modelled in VenSym. The code then generates a graph. This code does NOT work in Colab.

```python
# Import the packages
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read file
f = open('test1.txt')
s = f.readlines()

# Assign lines in the file to strings. By default, VenSym will output time in the first line
x = s[0]
y = s[1]

# Split strings and save values into lists
xs = x.split()
ys = y.split()

# Lists for numerical values
t = list()
v = list()
# Parse out text
for val in xs:
    try:t.append(int(val))
    except:continue
for val in ys:
    try:v.append(float(val))
    except:continue

# Save to dataframe
ts = pd.DataFrame(list(zip(t,v)))
ts.columns = ['Time','Profit'] # Change names as needed
```

```python
# PARAMETER SELECTION: Customize as needed here
# Resolution and figure size
sdpi = 300
plt.figure(figsize =(8,5),layout='constrained')
# Line properties
col = '0.5' # Color. For colorblindness, try 0.0 to 1.0
style = '-' # Options include '--', ':', '-.'
width = 3
# Marker properties
shape = 'o'# Marker shape, options include ' ','x','d','*'
size = 5

# Generate plot
plt.plot(ts['Time'],ts['Profit'], color=col, linestyle=style, linewidth=width, marker=shape, markersize=size)
plt.grid(True)

# Set X and Y axis range. Can be fixed to allow comparison between runs
plt.ylim(min(v),max(v)+1000000)
plt.xlim(min(t),max(t))

# Axis labels
plt.ylabel('Profit (Baht)',fontsize=12)
plt.xlabel('Time (months)',fontsize=12)

#Axis tick labels
plt.tick_params(labelsize = 12)

# Plot title
#plt.title('Graph')

# To save figure
#plt.savefig('Graph_name.png',dpi=sdpi) # Other format options include .tiff, .jpeg
```
